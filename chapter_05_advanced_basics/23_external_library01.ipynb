#### `sympy`
- 방정식 기호(symbol) 사용 기능

```python
from fractions import Fraction
import sympy
```

```python
# 가진 돈의 2/5로 학용품을 사는데 쓴 돈이 1,760원이다. 전체 돈은 얼마일까?
x = sympy.symbols("x")

# Fraction(x, y) == x / y
# sympy.Eq(a, b): a == b
f_x = sympy.Eq(x * Fraction(2, 5), 1760)

# sympy.solve(sympy.Eq() obj): Eq 방정식을 풀어 미지수의 값을 반환
result = sympy.solve(f_x)
print(result)
```

```python
# 남은 돈은 얼마일까?
remains = result[0] - 1760
print(remains)
```

```python
# x ** 2 = 1, 2차 방정식의 해 구하기
x = sympy.symbols("x")

f_x = sympy.Eq(x ** 2, 1)

# n차 방정식의 해는 n개의 item을 가진 리스트
result_x = sympy.solve(f_x)
print(result_x)
```

```python
# x + y = 10, x - y = 4, 연립방정식의 해 구하기
x, y = sympy.symbols("x,y")

f_plus = sympy.Eq(x + y, 10)
f_minus = sympy.Eq(x - y, 4)

# 미지수 2개 이상이면 결과값은 딕셔너리
result = sympy.solve([f_plus, f_minus])
print(result)
```

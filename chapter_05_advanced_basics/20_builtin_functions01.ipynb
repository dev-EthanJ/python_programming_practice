```python
# filter(func, iterable): iterable 객체 순서대로 func를 호출했을 떄 리턴값이 참인것만 묶어서 리턴
def positive(x):
    if x > 0:
        return x
    else:
        return 0
    
number_list = [1, -3, 2, 0, -5, 6]
tuple(filter(positive, number_list))
```

```python
tuple(filter(lambda x: x > 0, number_list))
```

```python
# isinstance(object, class): object가 해당 class의 객체인지 판단, 맞으면 return True, 아니면 return False
class Person:
    pass

class Student(Person):
    pass

jane = Person()
john = Student()
me = "human"

print(isinstance(jane, Person))
print(isinstance(john, Student))
print(isinstance(me, int))
```

```python
# 상속된 클래스의 경우 부모클래스인지 물어보면 return True
print(isinstance(john, Person))
```

```python
# len(object): 객체의 item 개수(길이) 리턴
print(len("python"))
print(len([1, 2, 3]))
print(len((1, 2, 3, 4)))
print(len({"a": 1}))
print(len(set((1, 2))))
```

```python
# list(iterable): iterable의 item을 묶은 list로 반환
print(list("python"))
print(list((1, 2, 3)))
print(list({"a": 1, "b": 2}))
```

```python
# list(list): 복사된 list를 리턴 > list 복사에 사용 가능
number_list = [1, 2 ,3]
new_list = list(number_list)
number_list[0] = 9
print(number_list, new_list)
```

```python
# min(iterable): iterable 객체의 최솟값을 리턴
min([1, 2, 3, 0])
```

```python
my_dict = {1: "I", 2: "my", 3: "me", 4: "mine"}
min(my_dict)
```

```python
reverse_dict = {}
for (key, value) in my_dict.items():
    reverse_dict[value] = key

print(reverse_dict)
print(min(reverse_dict))
```

```python
# zip(*iterable): 동일한 len을 가진 객체들의 item을 순서대로 묶어서(tuple) 리턴
my_list = [1, 2, 3]
my_str = "him"
my_dict = {"A": 90, "B": 80, "C": 70}

print(list(zip(my_list, my_str)))
print(list(zip(my_str, my_dict)))
print(list(zip(my_list, my_str, my_dict)))
```

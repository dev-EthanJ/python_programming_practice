#### `math.lcm()`
- 최소 공배수 (least common multiple) 구하기

```python
# 버스가 15분 간격, 기차가 25분 간격으로 올때 동시에 도착하는 시간은?
bus_gap = 15
train_gap = 25
math.lcm(bus_gap, train_gap)
```

```python
# 10분 간격으로 오는 택시도 추가하면, 동시에 도착하는 시간은?
taxi_gap = 10
math.lcm(bus_gap, train_gap, taxi_gap)
```

#### `random`

```python
# random.random(): 0.0 - 1.0 사이의 실수 리턴
import random

random.random()
```

```python
# random.randint(start, end): start 이상 end 이하 random int 리턴
random.randint(10, 20)
```

```python
# 무작위로 pop하는 함수
def random_pop(data):
    index = random.randint(0, len(data) - 1)
    return data.pop(index)

data = [1, 2, 3, 4, 5]
while data:
    print(random_pop(data))
```

```python
# random.choice(object): 객체의 무작위 item 하나를 선택하여 리턴
def new_randpop(data):
    item = random.choice(data)
    data.remove(item)
    return item

data = list(range(10, 16))
while data:
    print(new_randpop(data))
```

```python
# random.sample(population, k): population 객체에서 k개수만큼 샘플 무작위 추출
age_list = list(range(10, 20))

print(random.sample(age_list, len(age_list)))
print(random.sample(age_list, 3))
```

#### `itertools.permutations()`
- `itertools.permutations(iterable, r)`
- iterable 객체에서 r개를 순서있게 나열한 순열을 리턴

```python
import itertools

card_list = [1, 2, 3, 4]
permu_list = list(itertools.permutations(card_list, 2))
print(permu_list)
```

```python
print('5가지 카드에서 2개를 순서있게 선택하면')
for (a, b) in permu_list:
     print(f"({a}, {b})", end=" ")
print("의 순열로 선택할 수 있습니다.")
print(f"총 순열의 경우의 수는 {len(permu_list)}입니다.")
```

#### `functools.reduce()`
- `functools.reduce(function, iterable)`
- function 함수를 iterable 객체의 item에 index를 증가시키며 누적 적용

```python
# functools.reduce() 없이 총합 구하기 구현
def total_sum(data):
    result = 0
    for num in data:
        result += num
    return result

my_data = [2, 3, 4, 5, 6]
result = total_sum(my_data)
print(result)
```

```python
# functools.reduce()로 총합 구하기 구현
import functools

# ((((2 + 3) + 4) + 5) + 6)
result = functools.reduce(lambda x, y: x + y, my_data)
print(result)

def add(a, b):
    return a + b

new_result = functools.reduce(add, my_data)
print(new_result)
```

```python
# functools.reduce()로 최댓값, 최솟값 구하기
num_list = [2, 1, 5, 3, 6, 4]

max_num = functools.reduce(lambda x, y: x if x > y else y, num_list)
min_num = functools.reduce(lambda x, y: x if x < y else y, num_list)

print(max_num, min_num)
```

#### `operator.itemgetter()`
- `sorted()`같은 함수의 key 매개변수에 적용하여 다양한 기준으로 정렬할 수 있도록 한다.

```python
students = [
    ("jane", 22, "A"),
    ("david", 26, "C"),
    ("ethan", 24, "B")
]
```

```python
from operator import itemgetter

# index 2 == 나이 순서대로 정렬
result = sorted(students, key=itemgetter(2))
print(result)
```

```python
my_students = []
name_list = ["jane", "david", "ethan"]
age_list = [22, 23, 24]
grade_list = ["A", "C", "B"]

for (name, age, grade) in zip(name_list, age_list, grade_list):
    my_students.append({"age": age, "name": name, "grade": grade})

print(my_students)
```

```python
# "grade" key 기준으로 정렬
result = sorted(my_students, key=itemgetter("grade"))
print(result)
```

```python
# 클래스 객체의 속성으로 적용: operator.attrgetter()
from operator import attrgetter

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade


students_obj = [
    Student('jane', 22, "A"),
    Student('dave', 32, "B"),
    Student('sally', 27, "B"),
    Student('ava', 21, "C")
]

result = sorted(students_obj, key=attrgetter("name"))
for std in result:
    print(std.name)
```

#### `traceback`
- 프로그램 실행 중 발생한 오류 추적

```python
import traceback

def zero_divide():
    return 1 / 0

def func():
    zero_divide()

def main():
    try:
        func()
    except:
        print("오류가 발생했습니다.")
        print(traceback.format_exc())

main()
```

#### `json`
- JSON 데이터 처리

```python
import json

with open("myinfo.json", encoding="utf-8") as f:
    # json.load(file_obj): JSON 파일 읽기
    data = json.load(f)

print(data)
```

```python
# json.dump(dict, file_obj): 딕셔너리 객체를 JSON 파일로 생성
data = {"name": "나미리", "birth":"0505", "age": 27}
with open("myinfo.json", "a", encoding="utf-8") as f:
    json.dump(data, f) 
```

```python
# json.dumps(obj): 파이썬 자료형 객체를 JSON 문자열로 리턴
json_data = json.dumps(data)
print(json_data)
```

```python
# json.loads(JSON string): JSON 문자열을 딕셔너리로 변환
print(json.loads(json_data))
```

```python
# ensure_ascii=False: 한글 문자열이 ASCII 코드 문자열로 변경되는 것 방지
new_data = {"name": "신짱구", "birth": "0306", "age": 5}
json_data = json.dumps(new_data, ensure_ascii=False)
print(json_data)
```

```python
json.loads(json_data)
```

```python
# indent=N: JSON 문자열 정렬
indented_data = json.dumps(new_data, ensure_ascii=False, indent=2)
print(indented_data)
```

```python
# 여러 자료형에 대해 JSON 문자열 변환 가능
print(json.dumps([1, 2, 3]))
print(json.dumps((4, 5, 6)))
```

#### `urllib`
- URL을 읽고 분석

```python
import urllib.request

def get_wikidocs(page):
    resource = "https://wikidocs.net/{}".format(page)

    # 자동화 프로그램으로 보이지 않도록 User-agent 헤더 생성
    resource = urllib.request.Request(
        resource,
        headers= {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0.0.0 Safari/537.36"
        ),
        "Accept": "text/html,application/xhtml+xml",
        "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
        "Referer": "https://wikidocs.net/",
        }
    )

    # urllib.request.urlopen(url string): HTTP 요청 보낸 후 응답 객체 http.client.HTTPResponse 생성
    with urllib.request.urlopen(resource) as s:
        with open(f"wikidocs_{page}.html", "wb") as f:
            # s.read()로 url 웹페이지 읽기
            f.write(s.read())
    with open(f"wikidocs_{page}.html", "rb") as f:
        print(f.read())

get_wikidocs("12")
```

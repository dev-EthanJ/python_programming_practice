#### `time`

```python
# time.time(): 현재 시간을 실수 형태로 리턴, 1970/1/1 0:0:0초 기준 지난 시간을 초 단위로 리턴
import time

time.time()
```

```python
# time.localtime(): time.time() 값을 이용해서 각종 날짜 형태를 가진 구조체로 바꾼 후 리턴, defalut = 현재 시각
my_time = time.time()
print(time.localtime(my_time))
print(time.localtime(my_time).tm_year)
print(time.localtime())
```

```python
# time.asctime(): time.localtime() 값을 이용해서 날짜 표현 문자열을 리턴, default = 현재 시각 
my_localtime = time.localtime(my_time)
print(time.asctime(my_localtime))
print(time.asctime())
```

```python
# time.ctime(): time.asctime(time.localtime(time.time())) 리턴
time.ctime()
```

```python
# time.strftime("format string", [time.localtime()]): 형식 문자열에 맞춘 값 리턴, defalut = 현재 시각
this_time = time.localtime(time.time())
print(time.strftime("오늘은 %Y년 %m월 %d일 입니다.", this_time))
print(time.strftime("지금은 %p %I:%M:%S 입니다.", this_time))
```

```python
time.strftime("%c")
```

```python
# time.sleep(number): number초 만큼 멈춤
for i in range(10):
    print(i)
    time.sleep(1.5)
```

#### `math.gcd()`

#### `math.lcm()`
- 최소 공배수 (least common multiple) 구하기

```python
# 버스가 15분 간격, 기차가 25분 간격으로 올때 동시에 도착하는 시간은?
bus_gap = 15
train_gap = 25
math.lcm(bus_gap, train_gap)
```

```python
# 10분 간격으로 오는 택시도 추가하면, 동시에 도착하는 시간은?
taxi_gap = 10
math.lcm(bus_gap, train_gap, taxi_gap)
```

#### `random`

```python
# random.random(): 0.0 - 1.0 사이의 실수 리턴
import random

random.random()
```

```python
# random.randint(start, end): start 이상 end 이하 random int 리턴
random.randint(10, 20)
```

```python
# 무작위로 pop하는 함수
def random_pop(data):
    index = random.randint(0, len(data) - 1)
    return data.pop(index)

data = [1, 2, 3, 4, 5]
while data:
    print(random_pop(data))
```

```python
# random.choice(object): 객체의 무작위 item 하나를 선택하여 리턴
def new_randpop(data):
    item = random.choice(data)
    data.remove(item)
    return item

data = list(range(10, 16))
while data:
    print(new_randpop(data))
```

```python
# random.sample(population, k): population 객체에서 k개수만큼 샘플 무작위 추출
age_list = list(range(10, 20))

print(random.sample(age_list, len(age_list)))
print(random.sample(age_list, 3))
```

#### `itertools.zip_longest()`
- `itertools.zip_lingest(*iterables, fillvalue=None)`
- `zip`함수와 동일하게 동작, iterable의 길이가 다르다면 fillvalue값을 채움

```python
students = ['원빈', '소희', '유타', '리쿠', '위시']
snacks = ['candy', 'chocolate', 'jelly']

list(zip(students, snacks))
```

#### `itertools.permutations()`
- `itertools.permutations(iterable, r)`
- iterable 객체에서 r개를 순서있게 나열한 순열을 리턴

```python
import itertools

card_list = [1, 2, 3, 4]
permu_list = list(itertools.permutations(card_list, 2))
print(permu_list)
```

```python
print('5가지 카드에서 2개를 순서있게 선택하면')
for (a, b) in permu_list:
     print(f"({a}, {b})", end=" ")
print("의 순열로 선택할 수 있습니다.")
print(f"총 순열의 경우의 수는 {len(permu_list)}입니다.")
```

#### `functools.reduce()`
- `functools.reduce(function, iterable)`
- function 함수를 iterable 객체의 item에 index를 증가시키며 누적 적용

```python
# functools.reduce() 없이 총합 구하기 구현
def total_sum(data):
    result = 0
    for num in data:
        result += num
    return result

my_data = [2, 3, 4, 5, 6]
result = total_sum(my_data)
print(result)
```

```python
# functools.reduce()로 총합 구하기 구현
import functools

# ((((2 + 3) + 4) + 5) + 6)
result = functools.reduce(lambda x, y: x + y, my_data)
print(result)

def add(a, b):
    return a + b

new_result = functools.reduce(add, my_data)
print(new_result)
```

```python
# functools.reduce()로 최댓값, 최솟값 구하기
num_list = [2, 1, 5, 3, 6, 4]

max_num = functools.reduce(lambda x, y: x if x > y else y, num_list)
min_num = functools.reduce(lambda x, y: x if x < y else y, num_list)

print(max_num, min_num)
```

#### `operator.itemgetter()`
- `sorted()`같은 함수의 key 매개변수에 적용하여 다양한 기준으로 정렬할 수 있도록 한다.

```python
students = [
    ("jane", 22, "A"),
    ("david", 26, "C"),
    ("ethan", 24, "B")
]
```

```python
from operator import itemgetter

# index 2 == 나이 순서대로 정렬
result = sorted(students, key=itemgetter(2))
print(result)
```

```python
my_students = []
name_list = ["jane", "david", "ethan"]
age_list = [22, 23, 24]
grade_list = ["A", "C", "B"]

for (name, age, grade) in zip(name_list, age_list, grade_list):
    my_students.append({"age": age, "name": name, "grade": grade})

print(my_students)
```

```python
# "grade" key 기준으로 정렬
result = sorted(my_students, key=itemgetter("grade"))
print(result)
```

```python
# 클래스 객체의 속성으로 적용: operator.attrgetter()
from operator import attrgetter

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade


students_obj = [
    Student('jane', 22, "A"),
    Student('dave', 32, "B"),
    Student('sally', 27, "B"),
    Student('ava', 21, "C")
]

result = sorted(students_obj, key=attrgetter("name"))
for std in result:
    print(std.name)
```

#### `zipfile`
- 여러 개의 파일을 zip형식으로 합치거나 해제할 때 사용

```python
import zipfile

f = zipfile.ZipFile("./zip/my_module.zip", "w")
f.write("my_mod01.py")
f.write("my_mod02.py")

f.close()
```

```python
f = zipfile.ZipFile("./zip/my_module.zip")
f.extractall(path="zip")
f.close()
```

```python
# 특정 파일만 해제하기
f = zipfile.ZipFile("zip/my_module.zip")
f.extract("my_mod01.py")
```

```python
# 파일 압축해서 묶기
# zipfile.ZipFile("path", "w", compression=4 options, compresslevel=1~9)

f = zipfile.ZipFile("zip/compressed_module.zip", "w", compression=zipfile.ZIP_LZMA, compresslevel=9)
f.write("my_mod01.py")
f.write("my_mod02.py")

f.close()
```

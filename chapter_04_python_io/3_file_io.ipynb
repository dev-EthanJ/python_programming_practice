#### 쓰기모드로 열어서 내용쓰기

```python
file = open('write_file.txt', 'w')
for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    file.write(data)
file.close()
```

#### 파일을 읽는 여러가지 방법

```python
# 1. readline() 함수 이용하기
file = open("write_file.txt", "r")
while True:
    line = file.readline()
    print(line)
    if not line: break # 읽어들인 값이 없으면 break
file.close()
```

```python
# 2. readlines() 함수 이용하기
file = open("write_file.txt", "r")
lines = file.readlines()
print(type(lines))
for line in lines:
    print(line)
file.close()
```

```python
file = open("write_file.txt", "r")
lines = file.readlines()
print(type(lines))
for line in lines:
    line = line.strip()
    print(line)
file.close()
```

```python
# 3. read() 함수 사용하기
file = open("write_file.txt", "r")
data = file.read()
print(data)

file.close()
```

```python
# 4. 파일 객체를 for문과 함께 사용하기 > for문을 통해 줄단위로 읽을 수 있다
file = open("write_file.txt")
for line in file:
    print(line)
file.close()
```

#### 파일에 새로운 내용 추가하기: `a`

```python
file = open("write_file.txt", "a")
for i in range(11, 20):
    line = f'{i}번째 줄입니다.\n'
    file.write(line)
file.close()
```

```python
file = open("write_file.txt", "r")
for line in file:
    print(line)
file.close()
```

#### `with`문과 함께 사용하기

```python
# with문은 open()과 close()를 자동으로 해준다
with open("write_file.txt", "r") as file:
    for line in file:
        print(line)
```

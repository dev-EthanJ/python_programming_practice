#### `print` 제대로 알기

```python
# " "or' '로 둘러싸인 문자열의 나열은 + 연산과 동일하다
print("Life" "is" "too" "short")
```

```python
print("Life" + "is" + "too" + "short")
```

```python
print('Life' 'is' 'too' 'short')
```

```python
# 문자열 띄어쓰기는 쉼표로 한다
print("Life", "is", "too", "short")
```

```python
# 한 줄에 결괏값 출력하기 end=' '
for i in range(10):
    print(i, end=' ')

# print함수의 end매개변수의 defalut=='\n'
```

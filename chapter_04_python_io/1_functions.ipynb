#### 가변개수 매개변수 *

```python
# 여러개의 입력값을 받는 함수
def add_many(*args):
    result = 0
    for i in args:
        result += i
    return result
```

```python
print(add_many(1, 2, 3, 4))
print(add_many(5, 6, 7, 8, 9, 10))
```

```python
# *args + 다른 parameter
def add_mul(choice, *args):
    if choice == 'add':
        result = 0
        for i in args:
            result += i
    if choice == 'mul':
        result = 1
        for i in args:
            result *= i
    return result
    
```

```python
print(add_mul('add', 1, 2, 3, 4, 5))
print(add_mul('mul', 1, 2, 3, 4, 5))
```

#### 키워드 매개변수 `kwargs`
- 매개변수는 `key1 = value1, key2 = value2, ...`형태로 전달
- `kwargs` 변수는 `{key1: value1, key2: value2, ...}` 형태의 딕셔너리

```python
def print_keyword(**kwargs):
    print(kwargs)
```

```python
print_keyword(a = 1)
```

```python
print_keyword(name = 'Jane', age = 21)
```

#### 매개변수에 초깃값 미리 설정하기

```python
def say_myself(name, age, man=True):
    print(f'나의 이름은 {name}입니다.')
    print(f'나이는 {age}세 입니다.')
    if man: print('남성입니다.')
    else: print('여성입니다.')
```

```python
say_myself('김가인', 27)
```

```python
say_myself('노나인', 34, True)
```

```python
say_myself('박다인', 34, False)
```

```python
# 초깃값이 정해진 매개변수는 항상 맨 뒤에 와야한다
# def introduce_myself(name, man=True, age): > SyntaxError
def introduce_myself(name, age = 20, man = True):
    print(f'이름은 {name}이고, {age}세 입니다.')
    if man: print('남성입니다.')
    else: print('여성입니다.')
```

```python
introduce_myself('최차인')
```

```python
introduce_myself('하하인', 37)
```

```python
introduce_myself('김이박', man=False)
```

#### 함수 안에서 선언한 변수의 효력 범위

```python
# 함수 안에서 선언한 변수는 함수 안에서만 존재한다
a = 1
def add_a(a):
    a += 1

add_a(a)
print(a)
```

#### 함수 안에서 함수 밖의 변수를 변경하는 방법

```python
# 1. return 사용하기
a = 1
def add_num(a):
    a += 1
    return a

print(add_num(4))
```

```python
# 2. global 명령어 사용하기
a = 2
def add_glob():
    global a
    a += 1

add_glob()
print(a)
```

#### `lambda` 예약어
- `함수_식별자 = lambda 매개변수1, 매개변수2, ... : 매개변수를_이용한_표현식`

```python
def add_def(a, b):
    return a + b

result = add_def(3, 4)
print(result)
```

```python
add_lam = lambda a, b: a + b

new_result = add_lam(5, 6)
print(new_result)
```

#### 집합 자료형의 특징
- 중복을 허용하지 않는다
- 순서가 없다. (unordered)
    - 데이터의 중복을 제거하기 위해 set자료형을 필터로 사용할 수 있다

```python
# indexing, slicing 불가
# 사용하기 위해서 list() 또는 tuple() 사용해야 함
num_set = set([1, 2, 3])
num_set
```

```python
list(num_set)[2]
```

```python
tuple(num_set)[0]
```

#### 교집합, 합집합, 차집합 구하기

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
```

```python

# 교집합 구하기 - set & set
s1 & s2
```

```python
# 교집합 구하기 - set.intersection(set)
s1.intersection(s2)
```

```python
# 합집합 구하기 - set | set
s1 | s2
```

```python
# 합집합 구하기 - set.union(set)
s1.union(s2)
```

```python
# 차집합 구하기 set - set
print(s1 - s2)
print(s2 - s1)
```

```python
# 차집합 구하기 set.difference(set)
print(s1.difference(s2))
print(s2.difference(s1))
```

#### 집합 자료형 set 관련 함수

```python
# 값 1개 추가하기 - set.add(item)
s1 = set([1, 2, 3])
s1.add(4)
s1
```

```python
# 값 여러 개 추가하기 - set.update(item)
s1 = set([1, 2, 3])
s1.update([4, 5, 6])
s1.update((7, 8, 9))
s1.update("Hi")
s1
```

```python
# 특정 값 제거하기 - set.remove(item)
s1.remove(2)
s1
```

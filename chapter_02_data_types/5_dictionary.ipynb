## Chapter 02. 자료형 (Data type)
### 02-5. 딕셔너리 자료형 (Dictionary)
- 딕셔너리는 순차적으로 요소값을 구하지 않고, Key를 통해 Value를 얻는다

#### 딕셔너리 만들기
- `{Key1: Value1, Key2: Value2, Key3: Value3, ...}`

```python
dit = {'name':'pey', 'phone':'010-1234-5678', 'birthday':'1118'}
dit
```

```python
a = {1: "hi"}
a
```

```python
a = {'a': [1, 2], 34: 'ad', 'dict':[3, 4]}
a
```

#### 딕셔너리 쌍(Key: Value) 추가, 삭제하기

```python
a = {1: 'a'}
a[2] = 'b'
a
```

```python
a['name'] = 'micle'
a
```

```python
a[3] = [1, 2, 3]
a
```

```python
a[4] = (1, )
a
```

```python
# 딕셔너리 item 삭제 del[Key]
del a[1]
a
```

```python
# 딕셔너리 생성시 주의사항
# Key는 고유한 값, 중복 Key값 설정 시 하나를 제외한 나머지는 무시됨
my_dict = {3: 'd', 1: 'a', 2: 'b', 1: 'c'}
my_dict
```

```python
# Key에 list는 사용 불가하지만, tuple은 사용 가능
# 상수값만 Key로 사용 가능
new_dict = {1: 'a', (2,): 'c', (3, 4):[3, 4]}
new_dict[(2, )]
```

#### 딕셔너리 관련 함수

```python
# Key list 만들기 - dict.keys()
a = {'name': 'John', 'phone': '010-1234-5678', 'birthday': '1225'}
a.keys()
```

```python
for k in a.keys():
    print(k)
```

```python
# dict.keys() 객체를 list로 변환 - list(dict.keys())
list(a.keys())
```

```python
# Value list 만들기 - dict.values()
a.values()
```

```python
# Key, Value 쌍 얻기 - dict.items()
a.items()
```

```python
# Key, Value 쌍 모두 지우기 - dict.clear()
a.clear()
a
```

```python
# Key로 Value 얻기 - dict.get(Key)
a = {'name':'John', 'phone':'010-1234-5678', 'birthday':'19980710'}
a.get('name')
```

```python
# dict.get(Key) == dict[Key]
a['name']
```

```python
# 존재하지 않는 Key값으로 불러오려고 할 때
# print(a['height']) > KeyError
print(a.get('height')) 
```

```python
# 존재하지 않는 Key일때 디폴트값 설정 - dict.get(Key, default)
a.get('height', 'N/A')
```

```python
# Key가 딕셔너리 안에 있는지 조사하기 - Key in dict
user = {'name':'pey', 'phone':'010-1234-5678', 'birthday': '0623'}
print('name' in user)
print('height' in user)
```

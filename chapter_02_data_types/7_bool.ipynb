#### 자료형의 참과 거짓 (bool)
- string, list, tuple, dictionary 등의 값이 비어있으면 `False`, 비어있지 않으면 `True`
- 숫자 0 == `False`, 숫자 0 제외 == `True`
- `None` == `False`

```python
num_list = [1, 2, 3, 4]
# num_list는 계속 pop되므로, 언젠가 빈 리스트 []가 되어 조건식이 False가 된다
while num_list:
    print(num_list.pop())
```

```python
if []:
    print("True")
else:
    print("False")
```

```python
if [1, 2, 3]:
    print("True")
else:
    print("False")
```

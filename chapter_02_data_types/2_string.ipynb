#### format 함수를 사용한 포매팅
- `"문자열".format()`

```python
# 숫자 바로 대입
"I eat {0} apples".format(3)
```

```python
# 문자열 대입
"I eat {0} apples.".format("five")
```

```python
# 변수 대입
number = 10
day = "three"
"I ate {0} apples, so I was sick for {1} days.".format(number, day)
```

```python
# 변수 이름으로 넣기, 초기화 필수
"I ate {number} bananas, so I was happy for {day} days.".format(number = 5, day = "two")
```

```python
# 인덱스와 이름 혼용해서 넣기
"I ate {0} apples, so I was high for {day} days.".format(3, day = "seven")
```

```python
# 왼쪽 정렬 `:<`
"{0:<10}".format("hi")
```

```python
# 오른쪽 정렬 :>
"{0:>10}".format("hi")
```

```python
# 가운데 정렬 :^
"{0:^10}".format("hi")
```

```python
# 공백 채우기
"{0:=^10}".format("hi")
```

```python
"{0:!<10}".format("hi")
```

```python
# 소수점 표현하기
y = 123.154245
"{0:0.4f}".format(y)
```

```python
"{0:010.4f}".format(y)
```

```python
# 문자 표현하기
"{{ and }}".format()
```

#### f 문자열 포매팅
- python 3.6 이상부터 사용가능

```python
# 문자열 앞에 f 접두사
name = "홍길동"
age = 30
f"나의 이름은 {name}이고, 나이는 {age}세 입니다."
```

```python
# 계산식 지원
f'나는 내년에 {age + 1}세가 됩니다.'
```

```python
# 딕셔너리 사용
d = {'name':'김가인', 'age':30}
f'나의 이름은 {d['name']}입니다. 나이는 {d['age']}세 입니다.'
```

```python
# 정렬
f'Hi. {"hello":<10}'
```

```python
f'Hi. {"Hello0":>10}'
```

```python
f'Hi. {'Hello':^10}'
```

```python
# 공백 채우기
f'Hi. {"Hello":=^10}'
```

```python
f'Hi. {"Hello":!<10}'
```

```python
say = "Hello"
f'Hi. {say:=^10}'
```

#### 문자열 관련 함수들

```python
# 문자 개수 세기 str.count(char)
a = "hobby"
a.count('b')
```

```python
# 인덱스 알려주기 str.find(char)
data = "Python is the best choice"
data.find('b')
```

```python
data.find('y')
```

```python
ch = 't'
data.find(ch)
```

```python
data.find('k')
```

```python
# 인덱스 알려주기 str.index(char)
data = "Life is too short"
data.index('t')
```

```python
# data.index('k')
# ValueError
```

```python
# 문자열 삽입 str.join(str)
",".join("abcd")
```

```python
# 리스트와 튜플도 매개변수 입력 가능
",".join(['a', 'b', 'c', 'd'])
```

```python
# 대문자로 변경 str.upper()
data = "Hi, How are you?"
data.upper()
```

```python
# 소문자로 변경 str.lower()
data.lower()
```

```python
# 왼쪽 공백(문자) 지우기 str.lstrip([chars])
data = f'{"hi":^12}'
print(data)
data.lstrip()
```

```python
star = f'{"hi":*^12}'
print(star)
star.lstrip('*')
```

```python
my_str = f'abcd{"hi"}acdb'
print(my_str)
print(my_str.lstrip("abdc"))
```

```python
# 오른쪽 공백(문자) 지우기 str.rstrip([chars])
print(data.rstrip())
print(star.rstrip('*'))
print(my_str.rstrip("abcd"))
```

```python
# 양쪽 공백(문자) 지우기 str.strip([chars])
print(data.strip())
print(star.strip('*'))
print(my_str.strip("abcd"))
```

```python
# 문자열 대체하기 str.replace(str)
life = "Life is too short"
life.replace("Life", "My arm")
```

```python
# 문자열 나누기 str.split([char])
life.split()
```

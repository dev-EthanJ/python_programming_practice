#### 리스트의 인덱싱과 슬라이싱

```python
# list indexing
a = [1, 2, 3]
a
```

```python
a[0]
```

```python
a[0] + a[2]
```

```python
a[-1]
```

```python
a = [1, 2, 3, ['a', 'b', 'c']]
```

```python
a[0]
```

```python
a[-1]
```

```python
a[3]
```

```python
a[3][1]
```

```python
a[-1][2]
```

```python
# 삼중 리스트 인덱싱
my_list = [1, 2, ['a', 'b', ["Life", "is"]]]
my_list[2][2][0]
```

```python
# list slicing
a = [1, 2, 3, 4, 5]
a[0:2]
```

```python
my_str = "12345"
my_str[0:2]
```

```python
a = [1, 2, 3, 4, 5]
b = a[:2]
c = a[2:]
print(b)
print(c)
```

```python
# 슬라이싱으로 새 리스트 만들기
a = [1, 2, 3, 4, 5]
new_list = a[1:3]
new_list
```

```python
# 중첩된 리스트의 슬라이싱
a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]
a[2:5]
```

```python
a[3][:2]
```

#### 리스트 연산하기

```python
# 리스트 더하기 +
a = [1, 2, 3]
b = [4, 5, 6]
a + b
```

```python
# 리스트 곱하기(반복하기) *
my_list = [1, 2, 3]
my_list * 3
```

```python
# 리스트 길이 구하기 (item 개수 구하기)
len(a)
```

```python
# 리스트 연산 오류
a = [1, 2, 3]
# TypeError
# a[2] + "hi"
str(a[2]) + "hi"
```

#### 리스트의 수정과 삭제


```python
# 리스트 수정
a = [1, 2, 3]
a[2] = 4
a
```

```python
# del 함수를 이용해 리스트 item 삭제
a = [1, 2, 3]
del a[1]
a
```

```python
# 슬라이싱 + del 함수
a = [1, 2, 3, 4, 5]
del a[2:]
a
```

#### 리스트 관련 함수

```python
# 리스트 item 추가하기 list.append(item)
a = [1, 2, 3, 4]
a.append(5)
a
```

```python
a.append([6, 7])
a
```

```python
# 리스트 정렬 list.sort()
a = [1, 4, 2, 3]
a.sort()
a
# 원본 리스트 변경!
```

```python
# 리스트 순서 뒤집기 list.reverse()
a = ['a', 'c', 'b', 'd']
a.reverse()
a
```

```python
# 인덱스 반환 list.index(item)
a = [1, 2, 3]
a.index(1)
```

```python
a.index(3)
```

```python
# ValueError
# a.index(7)
```

```python
# 리스트 item 삽입 list.insert(index, item)
a = [1, 2, 3]
a.insert(1, 4)
a
```

```python
a.insert(3, [4, 5])
a
```

```python
# 리스트 item 제거 list.remove(item)
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
a
```

```python

a.remove(3)
a
```

```python
# 리스트 item 꺼내기 list.pop(index)
a = [1, 2, 3]
print(a.pop()) # pop() = pop(-1)
print(a)
```

```python
a = [1, 2, 3]
a.pop(1)
a
```

```python
# 리스트 item 개수 세기 list.count(item)
a = [1, 2, 3, 1, 2]
a.count(1)
```

```python
# 리스트 확장 list.extend(list)
a = [1, 2, 3]
a.extend([4, 5])
a
```

```python
b = [6, 7]
a.extend(b)
a
```

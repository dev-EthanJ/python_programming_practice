## Chapter 02. 자료형 (Data types)
### 02-8. 자료형의 값을 저장하는 공간, 변수 (variable)

#### 변수 만들기

```python
# a, b, c는 변수다!
a = 1
b = "python"
c = [1, 2, 3]
```

```python
my_var = [1, 2, 3]
# 1. [1, 2, 3] list data 객체가 메모리에 생성됨
# 2. my_var 식별자 변수는 [1, 2, 3] 메모리의 주소를 가리킴

# 변수가 가리키는 객체의 주소값 리턴하는 함수 - id(var)
print(id(my_var))
```

#### 리스트 복사

```python
a = [1, 2, 3, 4]
print(a)
id(a)
```

```python
# = 연산자는, 가리키는 객체의 주소값을 대입한다
b = a
id(b)
```

```python
# var is var - 동일한 객체를 가리키고 있는지 판단
a is b
```

```python
a[1] = 5
print(a)
print(b)
```

```python
# 같은 객체값을 가지지만 다른 주소를 가리키는 변수 만들기 (복사)
# 1. [:] 이용하기
my_list = [1, 2, 3, 4]
new_list = my_list[:]
my_list[0] = 0
print(my_list, new_list)
```

```python
# 2. copy 모듈 이용하기 - var = copy(var)
from copy import copy

a = [2, 4, 6]
b = copy(a)
print(a, b)
```

```python
b is a
```

#### 변수를 만드는 여러가지 방법

```python
# 튜플로 값 대입
a, b = (1, 2)
print(a, type(a))
print(b, type(b))
```

```python
(a, b) = "python", "life"
print(a, type(a))
print(b, type(b))
```

```python
# 튜플 괄호 생략 가능
a, b = 1, 'a'
(c, d) = ([1, 2], (3, 4))
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

```python
# 리스트로 변수 만들기
[age, height] = [25, 178]
print(age, type(age))
print(height, type(height))
```

```python
# 값 교환
a = 7
b = 2
print(a, b)
a, b = b, a
print(a, b)
```
